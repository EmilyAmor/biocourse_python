```python
import pandas
```

```python
oceania_data = pandas.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')
```

```python
oceania_data
```

```python
oceania_data.ix['Australia'].plot() #this is from pandas thus not needed to use plt, pandas knows
plt.xticks(rotation=45) #xticks are all settings for x-axis lables
```

```python
oceania_data.plot()
#ugh! ugly graph
```

```python
oceania_data.T.plot() #changes comlumns and rows around in data
plt.xticks(rotation=45)
plt.ylabel('GDP per capita')
```

```python
plt.style.use('ggplot') #ggplot is R ploting package
oceania_data.T.plot(kind='bar')
```

```python
years = []
for col in oceania_data.columns:
    year = col[-4:]
    years.append(year)
    
gdp_australia = oceania_data.ix['Australia'].tolist() #
gdp_nz = oceania_data.ix['New Zealand'].tolist()

plt.plot(years, gdp_australia, 'b-', label='Australia')
plt.plot(years, gdp_nz, 'g-', label = 'New Zealand')

plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita')
```

```python
plt.scatter(gdp_australia, gdp_nz)
plt.xlabel('Australia')
plt.ylabel('New Zealand')
```
